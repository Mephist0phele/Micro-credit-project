```python
df1=df.drop("msisdn",axis=1)
```

```python
df2=df1.drop("pcircle",axis=1)
```

```python
import warnings
warnings.filterwarnings('ignore')
df2["pdate"]=pd.to_datetime(df2['pdate'])
df2["Month"]=df2['pdate'].apply(lambda x:x.month)
df2["Day"]=df2["pdate"].apply(lambda x:x.day)
```

```python
df2.isnull().sum()
```

```python
df2.head()
```

```python
df3=df2.drop("pdate",axis=1)
```

```python
df3.head()
```

```python
df3.dtypes
```

```python
 #removing the outliers
from scipy.stats import zscore
zscore=abs(zscore(df3))
print(df3.shape)
df4=df3.loc[(zscore<3).all(axis=1)]
print(df4.shape)
```

```python
 #Dealing with the skewness
import numpy as np
for col in df4.columns:
    if df4.skew().loc[col]>0.55:
        df4[col]=np.log1p(df4[col])
```

```python
df3.isnull().sum()
```

```python
#df5=df4.replace(np.nan, 0)
```

```python
x=df3.drop("label",axis=1)
```

```python
x.isnull().sum()
```

```python
y=df3["label"]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
model= [LogisticRegression(), GaussianNB(), SVC(), DecisionTreeClassifier()]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=20,test_size=5)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
predlr=lr.predict(x_test)
print(accuracy_score(y_test,predlr))
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

```python
gb=GaussianNB()
gb.fit(x_train,y_train)
gb.score(x_train,y_train)
predlr=gb.predict(x_test)
print(accuracy_score(y_test,predlr))
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```
